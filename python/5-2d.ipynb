# Grafice continue bidimensionale

În exemplele de mai jos, vom prezenta cîteva dintre funcțiile ce se folosesc pentru reprezentări continue bidimensionale. Mai precis, vom reprezenta grafic funcții reale. Majoritatea exemplelor vor folosi funcții deja cunoscute din laboratorul de [grafice discrete](https://github.com/adimanea/fsa-softmat/blob/main/python/3-grafice-discrete.ipynb).

Nu uitați, dacă o anumită bibliotecă nu este recunoscută de Python, instalați-o cu `pip install {bibliotecă}`.

```python
# grafic simplu, cu puncte alese uniform
import numpy as np 
import matplotlib.pyplot as plt

# funcția de reprezentat, f(x) = x^2 & exp(-x^2)
def f(x):
    return x**2 * np.exp(-x**2)

# punctele alese ca mostre (100 puncte între 0 și 3)
x = np.linspace(start = 0, stop = 3, num = 100)

plt.plot(x, f(x))
plt.show()
```

Putem desena mai multe grafice pe aceeași figură foarte simplu. Copiem codul de mai sus și mai adăugăm o funcție `g(x) = x * exp(-x)`.

```python
# două grafice pe aceeași figură
import numpy as np 
import matplotlib.pyplot as plt

# funcția de reprezentat, f(x) = x^2 & exp(-x^2)
def f(x):
    return x**2 * np.exp(-x**2)

# a doua funcție
def g(x):
    return x * np.exp(-x)

# punctele alese ca mostre pentru f (100 puncte între 0 și 3)
xf = np.linspace(start = 0, stop = 3, num = 100)

## mostrele pentru g (pot fi diferite)
xg = np.arange(start = 0., stop = 3., step = 0.05)

# desenul pentru f
plt.plot(xf, f(xf))
# desenul pentru g, cu linie de culoare roșie
plt.plot(xg, g(xg), 'r-')

# arată-le pe amîndouă
plt.show()
```

## Formatarea graficului

În exemplul de mai jos, adăugăm următoarele opțiuni:
- alegerea limitelor pentru valorile de pe axe;
- adăugarea legendei, cu opțiunea de formatare matematică folosind cod LaTeX;
- adăugarea transparenței la linii;
- schimbarea tipului de linie

Desenele se vor baza pe funcțiile de mai sus.

```python
# opțiuni estetice
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 * np.exp(-x**2)

def g(x):
    return x * np.exp(-x)

xf = np.linspace(start = 0, stop = 3, num = 100)
xg = np.arange(start = 0., stop = 3., step = 0.05)

plt.plot(xf, f(xf), linestyle=':')
plt.plot(xg, g(xg), 'r-', alpha=0.3) # alpha = transparență, în procente

# formatarea axelor
plt.axis([0, 6, -0.05, 0.6]) # [xmin, xmax, ymin, ymax]

# legendă și titlu
plt.xlabel('axa OX')
plt.ylabel('axa OY')
plt.legend([r'$ f(x) = x \cdot \exp(-x^2)$', r'$ g(x) = x \cdot \exp(-x) $']) # r'string' = raw string = nu interpreta simbolurile speciale
# opțiunea 'r' (raw) face ca backslash-ul și alte simboluri să fie interpretate literal
plt.title('Două funcții pe același grafic')

plt.show()
```

## Exercițiu rezolvat: Punctele de intersecție între două grafice

Se dau funcțiile `f(x) = sin(x)` și `g(x) = cos^2(x)`. Găsiți punctele de intersecție ale celor două grafice, din intervalul `[0, 2*pi]`.

```python
import numpy as np
import matplotlib.pyplot as plt

# listă de valori între 0 și 2*pi
x = np.arange(0, 2*np.pi, step=0.05)

# [x1, x2, x3, x4, ...]
# [f(x1), f(x2), f(x3), f(x4), ...]
# => puncte (x1, f(x1)), (x2, f(x2)), ...

# funcțiile, salvate direct ca liste
f = np.sin(x)
g = np.cos(x) ** 2

plt.axis([0, 6.5, -1.1, 1.1]) # [xmin, xmax, ymin, ymax]

plt.plot(x, f, '-')
plt.plot(x, g, '-')

plt.xlabel(r'$ x \in [0, 2 \pi] $')
plt.legend([r'$ f(x) = \sin(x) $', r'$ g(x) = \cos^2(x) $'])

# caută punctele de intersecție și returnează ca listă
abscise = np.argwhere(np.diff(np.sign(f - g))).flatten()

print(f'Lista de abscise = {abscise}')

# argwhere = argument (abscisa = x) where ...
# dacă f > g pe intervalul (x1, x2) => f(x) > g(x) pentru x în intervalul (x1, x2) => f(x) - g(x) > 0
# dacă f < g pe intervalul (x1, x2) => f(x) < g(x) pentru x în intervalul (x1, x2) => f(x) - g(x) < 0
# dacă f = g în punctul xn => f(xn) = g(xn) => f(xn) - g(xn) = 0

# flatten([(1,2), (3,4)]) = [1,2,3,4]

# arată-le pe grafic, cerculețe roșii
plt.plot(x[abscise], f[abscise], 'ro')
plt.show()

# scrie-le
print("Punctele de intersecție dintre cele două grafice sînt:")
for a in abscise:
   print(f'({x[a]}, {f[a]})')
```
